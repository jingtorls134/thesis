```python
from mp_api.client import MPRester

api_key = input("Please enter your Materials Project API key: ")
```

```python
with MPRester(api_key) as mpr:
    print(mpr.materials.available_fields)
```

```python
docs = mpr.summary.search(fields=['formula_pretty', 'symmetry', 'thermo_type', 'uncorrected_energy_per_atom',
                                'energy_per_atom', 'energy_uncertainy_per_atom', 'formation_energy_per_atom', 
                                'energy_above_hull', 'is_stable', 'band_gap', 'cbm', 'vbm', 'efermi', 'is_gap_direct', 'is_metal'
                                ]
                        )
```

```python
docs
```

```python
import pandas as pd

data_dict_list = []

for doc in docs:

    data_dict = {
        'formula_pretty': doc.formula_pretty,
        'crystal_system': doc.symmetry.crystal_system.value,
        'symbol': doc.symmetry.symbol,
        'number': doc.symmetry.number,
        'point_group': doc.symmetry.point_group,
        'symprec': doc.symmetry.symprec,
        'version': doc.symmetry.version,
        'uncorrected_energy_per_atom': doc.uncorrected_energy_per_atom,
        'energy_per_atom': doc.energy_per_atom,
        'formation_energy_per_atom': doc.formation_energy_per_atom,
        'energy_above_hull': doc.energy_above_hull,
        'is_stable': doc.is_stable,
        'band_gap': doc.band_gap,
        'cbm' : doc.cbm,
        'vbm' : doc.vbm,
        'efermi': doc.efermi,
        'is_gap_direct': doc.is_gap_direct,
        'is_metal': doc.is_metal
    }

    data_dict_list.append(data_dict)

df = pd.DataFrame(data_dict_list)

```

```python
from matminer.featurizers.conversions import StrToComposition

stc = StrToComposition()

df = stc.featurize_dataframe(df, "formula_pretty")

df.head()
```

```python
from matminer.featurizers.composition import ElementFraction

ef = ElementFraction()

df = ef.featurize_dataframe(df, "composition")

df.head()
```
